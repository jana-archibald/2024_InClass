## Out of all the people who buy 36 cups of coffeethis month, 70% of them will win more than ***% of the time.

```r
normalplot<-function(m,sd,region=0){
  x<-seq(m-(3.5)*sd,m+(3.5)*sd,length=1000)
  y<-dnorm(x,m,sd)
  plot(x,y,type="l",xlab="",ylab="",ylim=c(0,max(y)*1.05),xaxs="i",yaxs="i",cex.axis = 2,yaxt='n', ann=FALSE)
  z<-x[x>region[1]]
  z<- z[z<region[2]]
  polygon(c(region[1],z,region[2]),
          c(0,dnorm(z,m,sd),0),col="gray")}
```
