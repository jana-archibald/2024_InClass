## More work with expected value!

What is the expected value of a 50 - 50 ticket?

1. An insurance policy costs 150 per year, and will pay policyholders 16,000 if they suffer a major injury (resulting in hospitalization) or 6000 if they suffer a minor injury (resulting in lost time from work). The company estimates that each year 1 in every 2000 policyholders will have a major injury and 1 in every 400 a minor injury. 

a. What is the company’s expected profit?

This work is also in the class notebook. Using standard multiply and add computations.

But I did want to share how I got R to check my work, for those of you who might want to use similar methods.

Here is the table for the probability distribution of the payouts.

```r
rbind(c("16000","6000",0),c(1/2000,1/400,1-1/2000-1/400))
```

Here is the list of payouts

```r
c(16000,6000,0)
```

This is the list of probablilities

```r
c(1/2000,1/400,1994/2000)
```

### The table above is the probability distribution for the amount of money paid out.  I can get R to compute the amount of profit in each situation by subtracting the payouts from 150.

(The quotes in the entry for the table are just to make things print nicely.)

150-list with payouts=profit for each situation.

```r
150-c(16000,6000,0)
```

To calculate the expected value I need to multiply each profit by its probability.

```r
c(1/2000,1/400,1994/2000)*(150-c(16000,6000,0))
```

Adding these up we can get the expected profit.

```r
-7.925-14.625+149.55
```

```r
sum(c(1/2000,1/400,1994/2000)*(150-c(16000,6000,0)))
```

Making Plots for discount cards

```r
card1<-c(0:20)*0
card1[1+1]<-0.9
card1[2+1]<-0.08
card1[20+1]<-0.02
barplot(card1,names.arg=c(0:20)*5,ylab="Probability",
        main="Plot for Discount Card 1")
```
