```r
normalplot<-function(m,sd,region=0){
  x<-seq(m-(3.5)*sd,m+(3.5)*sd,length=1000)
  y<-dnorm(x,m,sd)
  plot(x,y,type="l",xlab="",ylab="", bty="n", yaxt="n")
  z<-x[x>region[1]]
  z<-z[z<region[2]]
  polygon(c(region[1],z,region[2]),
          c(0,dnorm(z,m,sd),0),col="gray")
  abline(v=m)
  abline(h=0)}
```

# The normal approximation of the binomial distribution.

You can edit n and p to plot different binomial distributions.

```r
n=100;
p=0.5;
barplot(dbinom(c(0:n),n,p),
       names.arg=c(0:n),
         xlab="Number of Successes",ylab="Probability",
        main=paste("Binomial Distribution with n=",n,",p=",p),space=0)
```

```r
n=10;
p=0.1;
barplot(dbinom(c(0:n),n,p),
       names.arg=c(0:n),
         xlab="Number of Successes",ylab="Probability",
        main=paste("Binomial Distribution with n=",n,",p=",p),space=0)
lines(seq(0,n,length=100),dnorm(seq(0,n,length=100)-0.5,n*p,sqrt(n*p*(1-p))),col="red",lwd=2)
```

# Using the Normal Approximation


## I plan to flip a coin 100 times, what is the probability that get fewer than 45 heads?


## Option 1: Use binomial distribution.

```r
pbinom(44,100,0.5)
```

## Option 2: Use the normal approximation.

First I need to check if the distirbution is approximately normal.

\# of expected successes = np =100*0.5=50


\# of expected failures = n(1-p)=100*0.5=50

Both of these are bigger than 10.

Second we need to compute the mean and standard deviation.

$\mu=np$

$\sigma=\sqrt{np(1-p)}$

```r
100*0.5
sqrt(100*0.5*(1-0.5))
```

```r
normalplot(50,5,c(0,45))
```
