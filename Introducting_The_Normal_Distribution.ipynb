## Plotting Normal Distributions

```r
normalplot<-function(m,sd,region=0){
  x<-seq(m-(3.5)*sd,m+(3.5)*sd,length=1000)
  y<-dnorm(x,m,sd)
  plot(x,y,type="l",xlab="",ylab="",ylim=c(0,max(y)*1.05),xaxs="i",yaxs="i",cex.axis = 1,yaxt='n', ann=FALSE)
  z<-x[x>region[1]]
  z<- z[z<region[2]]
  polygon(c(region[1],z,region[2]),
          c(0,dnorm(z,m,sd),0),col="gray")}
```

```r
normalplot(100,15)
```

## Scores on an IQ test are approximately normally distributed with a mean of 100 and a standard deviation of 15.

## What is the probability that a person selected at random has ...

### ... an IQ less than 80

```r
normalplot(100,15,c(0,80))
```

```r
pnorm(80,100,15)
```

$$P(X<80)=pnorm(80,100,15)=0.09121...$$

### ... an IQ above  92

```r
normalplot(100,15,c(92,1000))
```

```r
1-pnorm(92,100,15)
```

(c)  $$P(X>92)=1-P(X<92)=1-0.296901428603851=0.703098571396149 $$

### ... an IQ between 100 and 130

```r
pnorm(130,100,15)-pnorm(100,100,15)
```

```r
normalplot(100,15,c(100,130))
```

(a)  $$P(100<X<130)=P(X<130)-P(X<100)=0.9772-0.5=0.4772$$

### ...  an IQ between  90 and 120

```r
normalplot(100,15,c(90,120))
```

```r
pnorm(120,100,15)-pnorm(90,100,15)
```

(d)  $$P(90<X<120)=P(X<120)-P(X<90)=0.9088-0.2525=0/6563$$
